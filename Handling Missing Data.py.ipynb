<h2>Handling Missing Data</h2>

<li>Parsing The Date Column from 'str' to 'Timestamp'</li><br></br>

```python
import pandas as pd 
df=pd.read_excel("sample_data.xls",header=2,parse_dates=['Date'])
type(df.Date[0])
df.set_index('Date',inplace=True)
df
```

<li>fillna() to replace some column values</li><br></br>

```python
df=pd.read_excel("sample_data.xls",header=2)
df
df.fillna({"Last Name":"Not Available","Gender":"Confidential"},inplace=True)
df
```

<li>Forward fill N.A values with previous values</li><br></br>

```python
df.fillna(method="ffill",inplace=True)
df
```

<li>Backward fill N.A values with previous values</li><br></br>

```python
df.fillna(method="bfill")
df
```

<li>Forward fill the NA values with limit fills </li><br></br>

```python
df.fillna(method="ffill",limit=1)
df
```

<li>Interpolate N.A values with between value</li><br>,</br>

```python
df.interpolate()
df
```

<li>Interpolate based on time values</li><br></br>

```python
df.interpolate(method="time")
df
```

<li>Drop all rows with N.A values</li><br></br>

```python
df.dropna()
```

</li>Drop all rows with all columns having N.A values</li><br></br>

```python
df.dropna(how="all")
```

<li>Drop all rows with aleast one column having N.A value</li><br></br>

```python
df.dropna(thresh=1)
```

<li>Reindexing all date values-(fill NA for all non-existent date values)</li><br></br>

```python
dt=pd.date_range("01/1/2017","01/11/2017")
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```
