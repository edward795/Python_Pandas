<h2>Reading from a python Dictionary</h2>

```python
dict={'day':['mon','tue','wed'],
     'temperature(in kelvins)':[221.2,174,213.4],
     'windspeed(in kmph)':[59,53,54],
     'event':['Rainy','Sunny','Winter']}

df2=pd.DataFrame(dict)
print(df2)
```

<h2>Reading from a python tuples-list</h2>

```python
tl=[('mon',32.5,'1/2/1017'),
   ('tue',34.6,'2/5/2017'),
   ('wed',45.7,'4/6/2017')]

df3=pd.DataFrame(tl,columns=['day','windspeed','date'])
df3
```

<h2>Reading from a python list of dictionaries</h2>

```python
li=[
    {'day':'mon','temperature':33,'windspeed':45,'event':'Rainy'},
    {'day':'tue','temperature':35,'windspeed':41,'event':'sunny'},
    {'day':'wed','temperature':39,'windspeed':51,'event':'Snowy'}
]
df4=pd.DataFrame(li)
df4
```
